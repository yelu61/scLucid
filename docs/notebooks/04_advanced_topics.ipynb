# Advanced Topics in scLucid

This notebook covers advanced features including differential expression, enrichment analysis, and performance optimization.

```python
import scanpy as sc
import scLucid
from scLucid.analysis import (
    find_markers,
    run_enrichment,
    DifferentialConfig,
    EnrichmentConfig
)
from scLucid.config import set_config
```

## Load Data

```python
adata = sc.read_h5ad("results/annotated.h5ad")
```

## Performance Optimization

```python
# Enable parallel processing
set_config(n_jobs=8)

# Enable caching for faster re-runs
from scLucid.qc import CacheConfig, enable_cache
cache_config = CacheConfig(
    enabled=True,
    cache_dir="~/.sclucid/cache"
)
enable_cache(config=cache_config)
```

## Advanced Differential Expression

```python
from scLucid.analysis import CompareGroupsConfig, compare_groups

# Compare two clusters
compare_config = CompareGroupsConfig(
    group1="cluster_0",
    group2="cluster_1"
)

results = compare_groups(adata, config=compare_config)
```

## Pathway Enrichment

```python
# Find markers first
markers = find_markers(adata, groupby="leiden")

# Run enrichment on top markers
enrichment_config = EnrichmentConfig(
    method="gsea",
    database="go_bp",
    max_padj=0.05
)

from scLucid.analysis import run_enrichment
enrichment_results = run_enrichment(
    markers,
    config=enrichment_config
)
```
